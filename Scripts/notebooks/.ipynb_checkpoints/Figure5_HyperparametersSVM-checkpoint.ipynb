```python
import seaborn as sns
import numpy as np
import scipy.io as io

#Import lighting object for shading surface plots.
from matplotlib.colors import LightSource


from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import axes3d

import matplotlib.pylab as pylab
import matplotlib.pyplot as plt
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
paper_rc = {'lines.linewidth': 0.4, 'lines.markersize': 1}                  
sns.set_context("paper", rc = paper_rc)
```

```python
#Set up parameters for the figures..
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (12, 6),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'axes.titleweight':'bold',
         'axes.labelweight': 'bold',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'
         }

pylab.rcParams.update(params)
pylab.rcParams["font.family"] = "serif"
pylab.rcParams["font.weight"] = "heavy"
pylab.rcParams['xtick.major.pad']='8'
pylab.rcParams['ytick.major.pad']='8'
```

```python
#Load the cross validation parameters here (this can be obtained from KernelParameterSelection by
#running classify_validationmeasure_dataset_1.m..
mat_svm1 = io.loadmat('/work/imagingQ/SpatialAttention_Drowsiness/microMeasuresAlertness_Neuroimage2018/'
                      'Scripts/mat_files/cv_parameters_svm1.mat')
data_svm1 = mat_svm1['cv_svm1']
mat_svm2 = io.loadmat('/work/imagingQ/SpatialAttention_Drowsiness/microMeasuresAlertness_Neuroimage2018/'
                      'Scripts/mat_files/cv_parameters_svm2.mat')
data_svm2 = mat_svm2['cv_svm2']
```

```python
xticks = np.linspace(-1,25,27)
yticks = np.linspace(-1,25,27)
xticlabels = (['2$^{-1}$','',
                '','','','','2$^{5}$'
                '','','','','',
                '','','','','','2$^{15}$'
                '','','','','',
                '','','','','','2$^{25}$'])
yticlabels = (['2$^{-5}$','','','','',''
                '','','','','2$^{-15}$'
                '','','','',''
                '','','','','','2$^{-5}$'
                '','','','','',''
                '','','','','','2$^{5}$'])
fig = plt.figure(figsize=(18,6))

z_val1 = data_svm1
X_val, Y_val = np.meshgrid(xticks, yticks)

Z= z_val1
X = X_val
Y = Y_val

# Normalize to [0,1]
Z = (Z-Z.min())/(Z.max()-Z.min())

colors = cm.viridis(Z)
rcount, ccount, _ = colors.shape

ax1 = fig.add_subplot(1, 2, 1, projection='3d')
surf = ax1.plot_surface(X, Y, Z, rcount=rcount, ccount=ccount,facecolors=colors, shade=True)
m = cm.ScalarMappable(cmap=cm.viridis)
m.set_array(Z)

plt.colorbar(m, shrink=0.5, aspect=5)
ticks = [-1, 5, 10, 15,20, 25]
xticlabels= (['2$^{-1}$','2$^{5}$','','2$^{15}$','','2$^{25}$'])
plt.xticks(ticks, xticlabels)
plt.yticks(ticks, xticlabels)

plt.tight_layout()
ax1.set(title='SVM#1',
       ylabel='C',
       xlabel='$\gamma$')
ax1.xaxis.labelpad = 20
ax1.yaxis.labelpad = 20

xticks = np.linspace(-1,5,7)
yticks = np.linspace(-1,5,7)
z_val2 = data_svm2
X_val, Y_val = np.meshgrid(xticks, yticks)

Z= z_val2
X = X_val
Y = Y_val
# Normalize to [0,1]
Z = (Z-Z.min())/(Z.max()-Z.min())

ax2 = fig.add_subplot(1, 2, 2, projection='3d')
surf = ax2.plot_surface(X, Y, Z, rcount=rcount, ccount=ccount,
                       facecolors=colors, shade=True)
m = cm.ScalarMappable(cmap=cm.viridis)
m.set_array(Z)
plt.colorbar(m, shrink=0.5, aspect=5)
ticks = [-1, 1, 3, 5]
xticlabels= (['2$^{-1}$','2$^{1}$','2$^{3}$','2$^{5}$'])
plt.xticks(ticks, xticlabels)
plt.yticks(ticks, xticlabels)

ax2.set(title='SVM#2',
       ylabel='C',
       xlabel='$\gamma$')
ax2.xaxis.labelpad = 20
ax2.yaxis.labelpad = 20
```
